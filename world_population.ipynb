```python
#importing Libraries

from bs4 import BeautifulSoup
import requests
```

```python
#Collecting Data from website

data_url = "https://www.worldometers.info/world-population/population-by-country/"
data_page = requests.get(data_url)
collect_data = BeautifulSoup(data_page.text,'html')
print(collect_data)
```

```python
#Prettify The data

print(collect_data.prettify())
```

```python
#Finding Table_data from website

table_data = collect_data.find_all('table')[0]
print(table_data)
```

```python
#Collecting Table Headers From Data

table_head=table_data.find_all('th')
print(table_head)
```

```python
#Converting The Table Headers as list

head_list =[att.text.strip() for att in table_head ]
```

```python
print(head_list)
```

```python
#Importing Library For Creating a DataFrame

import pandas as pd
Population_data = pd.DataFrame(columns = head_list)
```

```python
Population_data
```

```python
#Collecting Column Data From Table

col_data = collect_data.find_all('tr')
print(col_data)
```

```python
#Stores data as DataFrame

for r in col_data[1:]:
    rows_data = r.find_all('td')
    individual_rowdata = [tb_insertd.text.strip() for tb_insertd in rows_data ]
    length = len(Population_data)
    Population_data.loc[length] = individual_rowdata
    
```

```python
#Importing Libraries for Representing the Data in a Visual format

import matplotlib.pyplot as plt
from matplotlib.figure import Figure
```

```python
#Adjusting the Figure Sie

plt.figure(figsize=(10,6))
```

```python
#Creating a Bar chart

plt.bar(Population_data['Country (or dependency)'],Population_data['Population (2023)'], color="skyblue")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Population of Each Country in 2023")
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
#Creating a Pie_Chart 

plt.figure(figsize=(8, 8))
plt.pie(Population_data["Population (2023)"], labels=Population_data["Country (or dependency)"], autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Global Population in 2023")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
#Creating a Histogram chart

plt.figure(figsize=(8, 6))
plt.hist(Population_data["Yearly Change"], bins=5, color='skyblue', alpha=0.7, label='Yearly Change')
plt.hist(Population_data["Net Change"], bins=5, color='green', alpha=0.7, label='Net Change')
plt.xlabel("Population Change")
plt.ylabel("Frequency")
plt.title("Distribution of Yearly and Net Changes")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
Population_data['Land Area (Km²)'] = Population_data['Land Area (Km²)'].str.replace(',', '').astype(int)
```

```python
Population_data['Density (P/Km²)'] = Population_data['Density (P/Km²)'].astype(int)
```

```python
Population_data['Migrants (net)'] = Population_data['Migrants (net)'].astype(int)
```

```python
Population_data['Fert. Rate'] = pd.to_numeric(Population_data['Fert. Rate'], errors='coerce')
Population_data['Fert. Rate'] = Population_data['Fert. Rate'].fillna(0).astype(int)
```

```python
#Creating a Scatter Plot

plt.figure(figsize=(10, 8))
plt.scatter(Population_data["Land Area (Km²)"], Population_data["Density (P/Km²)"], s=Population_data["Land Area (Km²)"]/1000, c=Population_data["Density (P/Km²)"], cmap='coolwarm', alpha=0.7, edgecolors='k', linewidths=1)
plt.xlabel("Land Area (Km²)")
plt.ylabel("Population Density (people per Km²)")
plt.title("Bubble Chart: Land Area vs Population Density")
plt.colorbar(label='Population Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Creating a Correlation Matrix

import seaborn as sns
plt.figure(figsize=(8, 6))
correlation_matrix = Population_data[['Fert. Rate', 'Migrants (net)']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap: Fertility Rate vs Net Migration")
plt.tight_layout()
plt.show()
```
